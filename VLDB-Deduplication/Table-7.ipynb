```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,12))
fig, ax = plt.subplots()
my_x_ticks = np.arange(9,14,1)
plt.xticks(my_x_ticks)

x = [9,13]
k1 = [115,114]
k2 = [109,96]
k3 = [103,96]
k4 = [221,204]
k5 = [203,141]
k6 = [157,141]

plt.plot(x,k1,'s-',color = 'r',label="SSD w/o dedup")
plt.plot(x,k2,'s-',color = 'g',label="SSD w/ dedup")
plt.plot(x,k3,'s-',color = 'b',label="SSD w/ dedup & optimized caching")

plt.plot(x,k4,'o-',color = 'm',label="HDD w/o dedup")
plt.plot(x,k5,'o-',color = 'y',label="HDD w/ dedup")
plt.plot(x,k6,'o-',color = 'k',label="HDD w/ dedup & optimized caching")

for a,b in zip(x,k1):
    plt.text(a, b+0.02, '%d' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,k2):
    plt.text(a, b-0.02, '%d' % b, ha='center', va= 'top',fontsize=9)
for a,b in zip(x,k3):
    plt.text(a, b-0.02, '%d' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,k4):
    plt.text(a, b+0.02, '%d' % b, ha='center', va= 'top',fontsize=9)
for a,b in zip(x,k5):
    plt.text(a, b+0.02, '%d' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,k6):
    plt.text(a, b+0.02, '%d' % b, ha='center', va= 'top',fontsize=9)

plt.xlabel("buffer pool size (GB)")
plt.ylabel("seconds")
plt.legend(loc = "best")
plt.show()

fig.savefig('figure-7.pdf',dpi=600,format='pdf')
```
