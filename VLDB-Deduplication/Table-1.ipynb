```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
my_x_ticks = np.arange(2, 7, 1)
plt.xticks(my_x_ticks)

x = [2,3,4,6]
k1 = [191,350,506,720]
k2 = [175,262,381,513]
k3 = [430,1112,1474,2209]
k4 = [425,639,962,1398]
plt.plot(x,k1,'s-',color = 'r',label="SSD w/o dedup")
plt.plot(x,k2,'s-',color = 'g',label="SSD w/ dedup & optimized caching")
plt.plot(x,k3,'o-',color = 'y',label="HDD w/o dedup")
plt.plot(x,k4,'o-',color = 'b',label="HDD w/ dedup & optimized caching")
for a,b in zip(x,k1):
    plt.text(a, b+10, '%d' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,k2):
    plt.text(a, b-10, '%d' % b, ha='center', va= 'top',fontsize=9)
for a,b in zip(x,k3):
    plt.text(a, b+10, '%d' % b, ha='center', va= 'bottom',fontsize=9)
for a,b in zip(x,k4):
    plt.text(a, b-10, '%d' % b, ha='center', va= 'top',fontsize=9)
plt.xlabel("numbers of models")
plt.ylabel("seconds")
plt.legend(loc=(1.05,0))
plt.show()

fig.savefig('figure-1.pdf',dpi=600,format='pdf')
```
