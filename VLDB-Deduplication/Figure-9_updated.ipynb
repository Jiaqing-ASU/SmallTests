```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import FormatStrFormatter
import numpy as np

# plt.figure(figsize=(8,12))
fontsize=58
s_fontsize=56

fig, ax = plt.subplots(figsize=(34,8), dpi=100, constrained_layout=True)
my_x_ticks = np.arange(8, 16, 1)
plt.xticks(my_x_ticks, fontsize=s_fontsize)
plt.yticks(fontsize=s_fontsize)

x = [8,10,15]
k1 = [786,762,720]
k2 = [710,594,513]
k3 = [638,580,513]
k4 = [8120,2264,2209]
k5 = [4921,1435,1398]
k6 = [1720,1435,1398]

plt.plot(x,k1,'s-',color = '#e41a1c',label="SSD w/o dedup", linewidth=4)
plt.plot(x,k2,'s-',color = '#ff7f00',label="SSD w/ dedup", linewidth=4)
plt.plot(x,k3,'s-',color = '#377eb8',label="SSD w/ dedup & optimized caching", linewidth=4)

plt.plot(x,k4,'o-',color = '#4daf4a',label="HDD w/o dedup", linewidth=4)
plt.plot(x,k5,'o-',color = '#a65628',label="HDD w/ dedup", linewidth=4)
plt.plot(x,k6,'o-',color = '#984ea3',label="HDD w/ dedup & optimized caching", linewidth=4)

# for a,b in zip(x,k1):
#     plt.text(a, b+100, '%d' % b, ha='center', va= 'bottom',fontsize=11)
# for a,b in zip(x,k2):
#     plt.text(a, b+100, '%d' % b, ha='center', va= 'top',fontsize=11)
# for a,b in zip(x,k3):
#     plt.text(a, b-100, '%d' % b, ha='center', va= 'bottom',fontsize=11)
# for a,b in zip(x,k4):
#     plt.text(a, b+100, '%d' % b, ha='center', va= 'top',fontsize=11)
# for a,b in zip(x,k5):
#     plt.text(a, b+100, '%d' % b, ha='center', va= 'bottom',fontsize=11)
# for a,b in zip(x,k6):
#     plt.text(a, b+100, '%d' % b, ha='center', va= 'top',fontsize=11)

plt.xlabel("buffer pool size (GB)", fontsize=fontsize)
plt.ylabel("seconds", fontsize=fontsize)
plt.yscale('log')
y_major = matplotlib.ticker.LogLocator(base=10, numticks=50)
ax.yaxis.set_minor_locator(y_major)
plt.yticks([500,750,1000,1500,2500,5000,7500,10000])
plt.tick_params(axis='y', which='minor')
ax.yaxis.set_major_formatter(FormatStrFormatter("%.0f"))
ax.yaxis.set_minor_formatter(FormatStrFormatter("%.0f"))

plt.legend(loc=(0.35,0.55), fontsize=20)
plt.show()

fig.savefig('table-2-visual.pdf',dpi=100,format='pdf')
```
